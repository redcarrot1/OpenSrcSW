```python
import csv
f = open('seoul_2021.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

max_temp = -999.0
min_temp = 999.0
maxDiff = 0.0
min_date = ''
max_date = ''
maxDiff_date = ''

for row in data :
    if row[-1]=='':
        row[-1]=-999
    if row[-2]=='':
        row[-2]=999
    row[-1]=float(row[-1])
    row[-2]=float(row[-2])

    if max_temp<row[-1]:
        max_date = row[0]
        max_temp = row[-1]
        
    if min_temp>row[-2]:
        min_date = row[0]
        min_temp = row[-2]

    if maxDiff<(row[-1]-row[-2]):
        maxDiff_date = row[0]
        maxDiff = row[-1]-row[-2]
f.close()
print('일교차가 가장 큰 날')
print('날짜', maxDiff_date)
print('최저기온이 가장 낮은 날')
print('날짜', min_date)
print('최고기온이 가장 높은 날')
print('날짜', max_date)
```
